## **Teste 1**

```python


# Importando bibliotecas necess√°rias
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from sklearn.metrics import confusion_matrix, classification_report
import os
import cv2


# Montando o Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Definindo o caminho para o dataset
data_dir_path = os.path.join('/content/drive/MyDrive', 'cell_images', 'cell_images')

# Verificando se o caminho existe
if not os.path.exists(data_dir_path):
    raise OSError(f"Diret√≥rio de dados n√£o encontrado: {data_dir_path}")

print(f"Diret√≥rio de dados: {data_dir_path}")


# Fun√ß√£o para carregar as imagens e r√≥tulos do dataset
def load_images_and_labels(data_dir, limit_per_class=None):
    images = []
    labels = []

    # Caminho para as pastas Parasitized e Uninfected
    parasitized_path = os.path.join(data_dir, 'Parasitized')
    uninfected_path = os.path.join(data_dir, 'Uninfected')

    # Carregue imagens parasitadas (limitado ao n√∫mero especificado)
    count_parasitized = 0
    for filename in os.listdir(parasitized_path):
        if count_parasitized >= limit_per_class:
            break

        if filename.endswith('.png'):
            img = Image.open(os.path.join(parasitized_path, filename))
            img_array = np.array(img)
            images.append(img_array)
            labels.append(1)  # 1 representa parasitized
            count_parasitized += 1

    # Carregue imagens n√£o infectadas (limitado ao n√∫mero especificado)
    count_uninfected = 0
    for filename in os.listdir(uninfected_path):
        if count_uninfected >= limit_per_class:
            break

        if filename.endswith('.png'):
            img = Image.open(os.path.join(uninfected_path, filename))
            img_array = np.array(img)
            images.append(img_array)
            labels.append(0)  # 0 representa uninfected
            count_uninfected += 1

    return np.array(images), np.array(labels)

# Carregue as imagens e r√≥tulos (limitado a uma quantidade imagens de cada classe porque o dataset tem mais de 27 000 imagen)
limit_per_class = 5000
images, labels = load_images_and_labels(data_dir_path, limit_per_class)


# Visualize algumas imagens
plt.figure(figsize=(10, 5))
for i in range(5):
    plt.subplot(1, 5, i + 1)
    plt.imshow(images[i])
    plt.title(f"Label: {labels[i]}")
    plt.axis('off')
plt.show()

# An√°lise Explorat√≥ria de Dados
# Distribui√ß√£o das classes
sns.countplot(x=labels)
plt.title('Distribui√ß√£o das Classes')
plt.show()

# Pr√©-processamento
# Codifica√ß√£o de r√≥tulos
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(labels)

# Divida os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(images, encoded_labels, test_size=0.2, random_state=42)

# Redimensionando todas as imagens para um tamanho fixo (224x224)
target_size = (224, 224)
X_train_resized = [cv2.resize(img, target_size) for img in X_train]
X_test_resized = [cv2.resize(img, target_size) for img in X_test]

# Convertendo para numpy arrays
X_train_resized = np.array(X_train_resized)
X_test_resized = np.array(X_test_resized)

# Normaliza√ß√£o (opcional, dependendo da sua escolha de modelo)
X_train = X_train / 255.0
X_test = X_test / 255.0

# Criando uma inst√¢ncia do LabelEncoder
label_encoder = LabelEncoder()
# Aplicando fit_transform aos r√≥tulos de treinamento
y_train_encoded = label_encoder.fit_transform(y_train)
# Aplicando transform aos r√≥tulos de teste
y_test_encoded = label_encoder.transform(y_test)


# Criando a arquitetura do modelo
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(1, activation='sigmoid')  # Camada de sa√≠da bin√°ria para classifica√ß√£o
])

# Compilando o modelo
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Exibindo a arquitetura do modelo
model.summary()

# Treinando o modelo
history = model.fit(X_train_resized, y_train_encoded, epochs=10, batch_size=32, validation_data=(X_test_resized, y_test_encoded))

# Avaliando o desempenho do modelo no conjunto de teste
test_loss, test_accuracy = model.evaluate(X_test_resized, y_test_encoded)
print(f'\nTest Accuracy: {test_accuracy * 100:.2f}%')

# Realizando previs√µes no conjunto de teste
y_pred = model.predict(X_test_resized)
y_pred_classes = np.round(y_pred)

# Calculando a matriz de confus√£o
conf_matrix = confusion_matrix(y_test_encoded, y_pred_classes)

# Visualizando a matriz de confus√£o
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Uninfected', 'Parasitized'], yticklabels=['Uninfected', 'Parasitized'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Exibindo o relat√≥rio de classifica√ß√£o
class_labels = ['Uninfected', 'Parasitized']
print(classification_report(y_test_encoded, y_pred_classes, target_names=class_labels))


# Visualiza√ß√£o do hist√≥rico de treinamento
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.show()

# Visualiza√ß√£o da perda
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.show()
```

```python
# Importando bibliotecas necess√°rias
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from sklearn.metrics import confusion_matrix, classification_report
import os
import cv2
from keras.preprocessing.image import ImageDataGenerator
from google.colab import drive

# Montando o Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Definindo o caminho para o dataset
data_dir_path = os.path.join('/content/drive/MyDrive', 'cell_images', 'cell_images')

# Verificando se o caminho existe
if not os.path.exists(data_dir_path):
    raise OSError(f"Diret√≥rio de dados n√£o encontrado: {data_dir_path}")

print(f"Diret√≥rio de dados: {data_dir_path}")

# Fun√ß√£o para carregar as imagens e r√≥tulos do dataset
def load_images_and_labels(data_dir, limit_per_class=None):
    images = []
    labels = []

    # Caminho para as pastas Parasitized e Uninfected
    parasitized_path = os.path.join(data_dir, 'Parasitized')
    uninfected_path = os.path.join(data_dir, 'Uninfected')

    # Carregue imagens parasitadas (limitado ao n√∫mero especificado)
    count_parasitized = 0
    for filename in os.listdir(parasitized_path):
        if count_parasitized >= limit_per_class:
            break

        if filename.endswith('.png'):
            img = Image.open(os.path.join(parasitized_path, filename))
            img_array = np.array(img)
            images.append(img_array)
            labels.append(1)  # 1 representa parasitized
            count_parasitized += 1

    # Carregue imagens n√£o infectadas (limitado ao n√∫mero especificado)
    count_uninfected = 0
    for filename in os.listdir(uninfected_path):
        if count_uninfected >= limit_per_class:
            break

        if filename.endswith('.png'):
            img = Image.open(os.path.join(uninfected_path, filename))
            img_array = np.array(img)
            images.append(img_array)
            labels.append(0)  # 0 representa uninfected
            count_uninfected += 1

    return np.array(images), np.array(labels)

# Carregue as imagens e r√≥tulos (limitado a uma quantidade imagens de cada classe porque o dataset tem mais de 27 000 imagen)
limit_per_class = 5000
images, labels = load_images_and_labels(data_dir_path, limit_per_class)

# Visualize algumas imagens
plt.figure(figsize=(10, 5))
for i in range(5):
    plt.subplot(1, 5, i + 1)
    plt.imshow(images[i])
    plt.title(f"Label: {labels[i]}")
    plt.axis('off')
plt.show()

# An√°lise Explorat√≥ria de Dados
# Distribui√ß√£o das classes
sns.countplot(x=labels)
plt.title('Distribui√ß√£o das Classes')
plt.show()

# Pr√©-processamento
# Codifica√ß√£o de r√≥tulos
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(labels)

# Divida os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(images, encoded_labels, test_size=0.2, random_state=42)

# Redimensionando todas as imagens para um tamanho fixo (224x224)
target_size = (224, 224)
X_train_resized = [cv2.resize(img, target_size) for img in X_train]
X_test_resized = [cv2.resize(img, target_size) for img in X_test]

# Convertendo para numpy arrays
X_train_resized = np.array(X_train_resized)
X_test_resized = np.array(X_test_resized)

# Augmentation e Normaliza√ß√£o
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Gerador de treinamento
train_generator = train_datagen.flow(X_train_resized, y_train, batch_size=32)

# Nova arquitetura do modelo
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    BatchNormalization(),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

# Compilando o modelo
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Exibindo a arquitetura do modelo
model.summary()

# Treinando o modelo com Augmentation
history = model.fit(train_generator, epochs=5, validation_data=(X_test_resized, y_test))

# Avaliando o desempenho do modelo no conjunto de teste
test_loss, test_accuracy = model.evaluate(X_test_resized, y_test)
print(f'\nTest Accuracy: {test_accuracy * 100:.2f}%')

# Realizando previs√µes no conjunto de teste
y_pred = model.predict(X_test_resized)
y_pred_classes = np.round(y_pred)

# Calculando a matriz de confus√£o
conf_matrix = confusion_matrix(y_test, y_pred_classes)

# Visualizando a matriz de confus√£o
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Uninfected', 'Parasitized'], yticklabels=['Uninfected', 'Parasitized'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Exibindo o relat√≥rio de classifica√ß√£o
class_labels = ['Uninfected', 'Parasitized']

# Criando gerador para o conjunto de teste sem aumento de dados
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow(X_test_resized, y_test, batch_size=32, shuffle=False)

# Realizando previs√µes no conjunto de teste
y_pred = model.predict(test_generator)
y_pred_classes = np.round(y_pred)

# Exibindo o relat√≥rio de classifica√ß√£o
print(classification_report(y_test, y_pred_classes, target_names=class_labels))

# Visualiza√ß√£o do hist√≥rico de treinamento
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.show()

# Visualiza√ß√£o da perda
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.show()

```
